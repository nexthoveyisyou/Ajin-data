```python
import pandas as pd
```

```python
s1 = pd.DataFrame([[10, 20], [30, 40], [50, 60], [70, 80]], columns=['수학', '과학'] )
s1
```

```python
s2 = pd.DataFrame([[5, 6], [13, 15]], columns=['과학', '수학'])
s2
```

```python
#axis가 붙이는 방향을 나타냄
pd.concat([s1, s2], axis=0)
```

```python
s3 = pd.DataFrame([[5, 87], [15, 90]], columns=['과학', '역사'])
s3
```

```python
# 겹치는 부분이 없을 경우 NaN으로 처리
pd.concat([s1, s3])
```

```python
#join옵션을 inner로 변경하면 겹치는 부분만 출력
pd.concat([s1, s3], join='inner')
```

```python
#가로 방향으로 붙이기 위한 데이터
s4 = pd.DataFrame([[36, 100], [13, 46], [57, 68], [26, 37]], columns=['역사', '코딩'], index=[1, 2, 3, 0])
s4
```

```python
#axis=1 옵션으로 가로 방향을 지정
pd.concat([s1, s4], axis=1)
```

```python
# 데이터의 크기가 다른 경우
s5 = pd.DataFrame([[36, 25], [13, 46], [57, 68]], columns=['역사', '코딩'], index=[2, 3, 4])
# s5.index = [1, 2, 3]
s5
```

```python
#inner 옵션을 주면 겹치지 않는 데이터는 삭제
pd.concat([s1, s5], axis=1, join='inner')
```

```python
# concat - index를 기준으로 한 merge와 같다.
# concat 가로 방향으로 붙일 때 index를 기준으로 ...
# merge 는 index말고 컬럼 지정 가능
# [Country LS]   [Country GDP]
```

# pivot, melt
 - 일반적인 관계형 데이터베이스(DBMS)에서는 다차원의 데이터를 2차원으로 조직화
 - 데이터를 조직화하는 과정에서 다양한 형태가 된다.
 - key의 역할을 하는 부분과 실질적인 value의 역할을 하는 데이터가 구분됨
  - 2차원으로 구성된 value -> melt -> 1차원
  - 1차원 -> pivot -> 2차원

```python
df = pd.concat([s1, s4], axis=1)
df['학번'] = [1, 2, 3, 4]
df
```

```python
melted = df.melt(id_vars='학번', value_vars=['수학', '과학', '역사', '코딩'], value_name='점수', var_name='과목')
#melted.columns = ['학번', '과목', '점수']
# melted.rename({'variable':'과목', 'value':'점수'}, axis=1, inplace=True)
melted
```

```python
pv = melted.pivot(index='학번', columns='과목', values='점수')
pv
pv.reset_index()
```

```python
students = pd.DataFrame({'학번':[1, 2, 3, 5],
                         '이름':['김', '이', '박', '성'],
                         '거주지':['수성', '수성', '화성', '토성']})
students
```

```python
melted.merge(students, on='학번', how='outer')
# pd.merge(melted, students, on='학번', how='right')
```
