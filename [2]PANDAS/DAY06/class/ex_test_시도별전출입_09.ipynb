```python
## 모듈 로딩
import pandas as pd 
```

```python
## 데이터 준비
DATA_FILE = '../DATA/시도별 전출입 인구수.xlsx'
```

[2] 데이터 로딩 <hr>

```python
## xlsx ==> DataFrame으로 로딩
df = pd.read_excel(DATA_FILE, sheet_name="데이터")
```

```python
## 앞부분 5개 행을 출력하세요.
print("[앞부분 5개 행]")
df.head()
```

```python
## 각 컬럼의 자료형을 확인하세요.
print(f"[컬럼별 자료형]")
df.dtypes
```

```python
## -------------------------------------------------------------
## 데이터 전처리 진행 체크 
## -------------------------------------------------------------
## 1. 전출지별 컬럼 => 이전값으로 채우면 결측치 처리
## 2. 0번행 의미 없는 행 => 삭제
## 3. 1970년 컬럼 ~ 2017년 컬럼 타입 object => '-' 문제
## -------------------------------------------------------------
```

[3] 기본 전처리<hr>

```python
## 1. 전출지별 컬럼 => 이전값으로 채우면 결측치 처리
df['전출지별'] = df['전출지별'].ffill()
```

```python
## 2. 0번행 의미 없는 행 => 삭제
df.drop(index=0, inplace=True)
```

```python
## 3. 1970년 컬럼 ~ 2017년 컬럼 타입 object => '-' 문제
df = df.mask(df=='-', 0)
df[df.columns[2:]] = df[df.columns[2:]].astype(int)
df.dtypes
```

---
#### 문제 9) 기초 통계 값을 확인하세요.
[요구사항]
1. 1970년부터 2017년까지 연도 컬럼만 선택하세요.
2. 각 연도별 이동자수 평균값을 구하세요.
3. 각 연도별 이동자수 최댓값을 구하세요.
4. 2017년 이동자수의 평균, 중앙값, 최댓값, 최솟값, 표준편차를 각각 출력하세요.
5. describe()를 사용하여 전체 연도 컬럼의 기술 통계 정보를 출력하세요.
---

```python
# 1. 1970년부터 2017년까지 연도 컬럼만 선택하세요.
## 특정 컬럼 선택 후 기본 정보 확인
yearDF = df[ df.columns[2:] ]

## 컬럼에 '-' 처리
yearDF = yearDF.mask(yearDF =='-', 0).astype(int)

## 기본 정보
yearDF.info()
```

```python
# 2. 각 연도별 이동자수 평균값을 구하세요.
print("[각 연도별 평균값]")
display(yearDF.mean())
```

```python
# 3. 각 연도별 이동자수 최댓값을 구하세요.
print("[각 연도별 최댓값]")
display(yearDF.max())
```

```python
# 4. 2017년 이동자수의 평균, 중앙값, 최댓값, 최솟값, 표준편차를 각각 출력하세요.
print("[2017년 통계]")
print("평   균 :", yearDF['2017'].mean())
print("중 앙 값 :", yearDF['2017'].median())
print("최 댓 값 :", yearDF['2017'].max())
print("최 솟 값 :", yearDF['2017'].min())
print("표준편차 :", yearDF['2017'].std())
```

```python
# 5. describe()를 사용하여 전체 연도 컬럼의 기술 통계 정보를 출력하세요.
display(yearDF.describe())
```
