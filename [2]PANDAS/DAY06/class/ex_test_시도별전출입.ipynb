---

```python
## 1. 1970년부터 2017년까지의 연도 컬럼을 선택하세요.
year_cols = df2.columns[2:]
year_cols
```

```python
## 특정 컬럼 선택 후 기본 정보 확인
yearDF = df2[ year_cols ]
display( yearDF.head(3) )
yearDF.info()
```

```python
## 2. 연도별 이동자수 데이터에 포함된 "-" 값을 특정값으로 변경하기
## -------------------------------------------------------------------------
## FutureWarning : 현재 코드는 실행되지만 앞으로 Pandas 버전 변경 시 변경 경고
## -------------------------------------------------------------------------
# C:\Users\Public\Documents\ESTsoft\CreatorTemp\ipykernel_31568\3585299255.py:2: FutureWarning: 
# Downcasting behavior in `replace` is deprecated and will be removed in a future version. 
# To retain the old behavior, explicitly call `result.infer_objects(copy=False)`. 
# To opt-in to the future behavior, set `pd.set_option('future.no_silent_downcasting', True)`
#   yearDF.replace("-", 0).astype('int')
# "-"를 0으로 바꾸는 과정 => Pandas가 내부적으로 자료형을 자동 추론
#                         여기서 자동 다운캐스팅 방식이 앞으로 바뀔 예정
# [해결]
# 가장 깔끔한 해결은 replace 후 명시적으로 infer_objects()를 호출
## -------------------------------------------------------------------------
yearDF = yearDF.replace("-", 0).infer_objects(copy=False).astype("int")
```

```python
## [권장] 둘 다 조건에 따라 값을 바꾸는 Pandas 메서드
## => mask(조건, 변경값) : 조건이 True인 경우 값 변경 
yearDF = yearDF.mask(yearDF == "-", 0)

## => where(조건, 변경값) : 조건이 True인 값은 그대로 두고, False인 값 변경
yearDF = yearDF.where(yearDF != "-", 0).astype("int")
```

```python
## 2. 연도별 이동자수 데이터에 포함된 "-" 값을 결측치로 처리할 수 있도록 변환하세요.
import numpy as np
yearDF = yearDF.replace("-",np.nan)
```

```python
## 3. 변환 후 연도 컬럼들의 자료형을 출력하세요.
yearDF.dtypes.head()
```

```python
## 4. 전체 결측치 개수를 출력하세요.
#print(f"[년도별 결측치 개수]\n{yearDF.isna().sum()}")
print(f"[전  체 결측치 개수]  {yearDF.isna().sum().sum()}개")
```

```python
# 데이터 준비
dataDF = df2.copy()

# 전출지별 결측치 채우기
dataDF['전출지별'] = dataDF['전출지별'].ffill()

# 1997 ~ 2017 '-' 처리
dataDF[dataDF.columns[2:]]= dataDF[dataDF.columns[2:]].mask(dataDF == '-', 0).astype(int)

dataDF.dtypes
```

```python
# 1. 전출지별이 "서울특별시"인 데이터만 추출한 후 복사본을 만드세요.
seoulDF = dataDF[ dataDF['전출지별'] == '서울특별시' ].copy()
```

```python

# 2. 2017년 이동자수에서 2010년 이동자수를 뺀 값을 "증감_2010_2017" 컬럼으로 추가하세요.
seoulDF["증감_2010_2017"] = seoulDF['2017'] - seoulDF['2010']
seoulDF.head(2)
```

```python
# 3. "증감_2010_2017" 값이 양수이면 "증가", 0이면 "동일", 음수이면 "감소"로 표시하는 "변화구분" 컬럼을 추가하세요.

# seoulDF["변화구분"] = np.where(
#     seoulDF["증감_2010_2017"] > 0, "증가",
#     np.where(seoulDF["증감_2010_2017"] == 0, "동일", "감소")
# )

seoulDF["변화구분"] = "동일"
seoulDF["변화구분"] = seoulDF["변화구분"].mask( seoulDF["증감_2010_2017"] > 0,  "증가" )
seoulDF["변화구분"] = seoulDF["변화구분"].mask(seoulDF["증감_2010_2017"] < 0,  "감소" )
```

```python
# 4. 전입지별이 "서울특별시"인 행의 변화구분 값을 "자기지역"으로 수정하세요.
seoulDF.loc[seoulDF["전입지별"] == "서울특별시", "변화구분"] = "자기지역"
```
