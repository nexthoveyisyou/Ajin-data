```python
## -------------------------------------------------------------
## 데이터 전처리 진행 체크 
## -------------------------------------------------------------
## 1. 전출지별 컬럼 => 이전값으로 채우면 결측치 처리
## 2. 0번행 의미 없는 행 => 삭제
## 3. 1970년 컬럼 ~ 2017년 컬럼 타입 object => '-' 문제
## -------------------------------------------------------------
```

[3] 기본 전처리<hr>

```python
## 1. 전출지별 컬럼 => 이전값으로 채우면 결측치 처리
df['전출지별'] = df['전출지별'].ffill()
```

```python
## 2. 0번행 의미 없는 행 => 삭제
df.drop(index=0, inplace=True)
```

```python
## 3. 1970년 컬럼 ~ 2017년 컬럼 타입 object => '-' 문제
df = df.mask(df=='-', 0)
df[df.columns[2:]] = df[df.columns[2:]].astype(int)
df.dtypes
```

#### 문제 7) query()를 사용하여 2017년 이동자수가 100000명 이상인 데이터를 추출하세요.

[요구사항]
1. 연도 컬럼을 숫자형으로 변환한 DataFrame을 사용하세요.
2. query()를 사용하여 2017년 이동자수가 100000명 이상인 행을 추출하세요.
3. 전출지별, 전입지별, 2017 컬럼만 출력하세요.
4. 2017년 이동자수 기준으로 내림차순 정렬하세요.

[주의사항]
- 컬럼명이 숫자 형태이면 query()에서 사용하기 어려울 수 있습니다.
- 필요한 경우 2017 컬럼명을 "Y2017"과 같이 변경한 후 query()를 사용하세요.
---

```python
## 1. 연도 컬럼을 숫자형으로 변환한 DataFrame을 사용하세요.
partDF = df.copy()
partDF[partDF.columns[2:]] = partDF[partDF.columns[2:]].replace("-", 0).astype("int")
```

```python
partDF[partDF.columns[2:]] = partDF[partDF.columns[2:]].mask(partDF[partDF.columns[2:]] == "-", 0).astype(int)
```

```python
partDF.dtypes.head(3)
```

```python
## 2. query()를 사용하여 2017년 이동자수가 100000명 이상인 행을 추출하세요.
## 2-1: 불린인덱싱
partDF[ partDF['2017'] >= 1000000 ].sort_index(axis=1, ascending=False)
```

```python
## 2-2. query()메서드
## -> query()에서 숫자 컬럼명을 바로 쓰기 어렵기 때문에 2017 컬럼명을 변경
#partDF = partDF.rename(columns={'2017': "Y2017"})

partDF = partDF.rename( columns={str(year): "Y" + str(year) for year in range(1970, 2018)} )
print(partDF.shape, partDF.columns, sep='\n')

partDF = partDF.query("Y2017 >= 100000")
partDF.shape
```

```python
## 3. 전출지별, 전입지별, 2017 컬럼만 출력하세요.
partDF = partDF[['전출지별','전입지별', 'Y2017']]
partDF.head()
```

```python
## 4. 2017년 이동자수 기준으로 내림차순 정렬하세요.
partDF = partDF.sort_values(by="Y2017", ascending=False).reset_index(drop=True)
partDF.head(3)
```
