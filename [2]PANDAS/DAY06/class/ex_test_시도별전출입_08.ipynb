```python
## 모듈 로딩
import pandas as pd 
```

```python
## 데이터 준비
DATA_FILE = '../DATA/시도별 전출입 인구수.xlsx'
```

[2] 데이터 로딩 <hr>

```python
## xlsx ==> DataFrame으로 로딩
df = pd.read_excel(DATA_FILE, sheet_name="데이터")
```

```python
## 앞부분 5개 행을 출력하세요.
print("[앞부분 5개 행]")
df.head()
```

```python
## 데이터의 행수와 열수를 출력하세요.
print(f"[행/열 개수]  {df.shape}")
print(f"행 : {df.shape[0]}개, 열 : {df.shape[1]}개")
```

```python
## 각 컬럼의 자료형을 확인하세요.
print(f"[컬럼별 자료형]")
df.dtypes
```

```python
## -------------------------------------------------------------
## 데이터 전처리 진행 체크 
## -------------------------------------------------------------
## 1. 전출지별 컬럼 => 이전값으로 채우면 결측치 처리
## 2. 0번행 의미 없는 행 => 삭제
## 3. 1970년 컬럼 ~ 2017년 컬럼 타입 object => '-' 문제
## -------------------------------------------------------------
```

[3] 기본 전처리<hr>

```python
## 1. 전출지별 컬럼 => 이전값으로 채우면 결측치 처리
df['전출지별'] = df['전출지별'].ffill()
```

```python
## 2. 0번행 의미 없는 행 => 삭제
df.drop(index=0, inplace=True)
```

```python
## 3. 1970년 컬럼 ~ 2017년 컬럼 타입 object => '-' 문제
df = df.mask(df=='-', 0)
df[df.columns[2:]] = df[df.columns[2:]].astype(int)
df.dtypes
```

---
#### 문제 8) 인덱스를 설정하고 정렬하세요.

[요구사항]
1. 전출지별과 전입지별 컬럼을 행 인덱스로 설정하세요.
2. 설정된 인덱스를 기준으로 오름차순 정렬하세요.
3. loc을 사용하여 전출지별이 "서울특별시"인 데이터만 선택하세요.
4. 선택 결과에서 2017년 이동자수 상위 5개 지역을 출력하세요.
---

```python
# 1. 전출지별과 전입지별 컬럼을 행 인덱스로 설정하세요.
dataDF2 = df.set_index( keys=['전출지별','전입지별'] )
dataDF2.index
```

```python
# 2. 설정된 인덱스를 기준으로 오름차순 정렬하세요.
dataDF2 = dataDF2.sort_index()
dataDF2.index
```

```python
# 3. loc을 사용하여 전출지별이 "서울특별시"인 데이터만 선택하세요.
#    멀티인덱스에서 level=0이 서울특별시
seoulDF = dataDF2.loc['서울특별시']
```

```python
# 4. 선택 결과에서 2017년 이동자수 상위 5개 지역을 출력하세요.
top5 = seoulDF.sort_values(by='2017', ascending=False).head(5)
display(top5[['2017']])
```
