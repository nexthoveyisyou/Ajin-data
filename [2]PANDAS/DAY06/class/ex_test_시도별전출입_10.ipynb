```python
## 모듈 로딩
import pandas as pd 
```

```python
## 데이터 준비
DATA_FILE = '../DATA/시도별 전출입 인구수.xlsx'
```

[2] 데이터 로딩 <hr>

```python
## xlsx ==> DataFrame으로 로딩
df = pd.read_excel(DATA_FILE, sheet_name="데이터")
```

```python
## 각 컬럼의 자료형을 확인하세요.
print(f"[컬럼별 자료형]")
df.dtypes
```

```python
## -------------------------------------------------------------
## 데이터 전처리 진행 체크 
## -------------------------------------------------------------
## 1. 전출지별 컬럼 => 이전값으로 채우면 결측치 처리
## 2. 0번행 의미 없는 행 => 삭제
## 3. 1970년 컬럼 ~ 2017년 컬럼 타입 object => '-' 문제
## -------------------------------------------------------------
```

[3] 기본 전처리<hr>

```python
## 1. 전출지별 컬럼 => 이전값으로 채우면 결측치 처리
df['전출지별'] = df['전출지별'].ffill()
```

```python
## 2. 0번행 의미 없는 행 => 삭제
df.drop(index=0, inplace=True)
```

```python
## 3. 1970년 컬럼 ~ 2017년 컬럼 타입 object => '-' 문제
df = df.mask(df=='-', 0)
df[df.columns[2:]] = df[df.columns[2:]].astype(int)
df.dtypes
```

### 문제 10) 행/열을 추가하고 값을 수정하세요.

[요구사항]
1. 전출지별이 "서울특별시"인 데이터만 추출한 후 복사본을 만드세요.
2. 2017년 이동자수에서 2010년 이동자수를 뺀 값을 "증감_2010_2017" 컬럼으로 추가하세요.
3. "증감_2010_2017" 값이 양수이면 "증가", 0이면 "동일", 음수이면 "감소"로 표시하는 "변화구분" 컬럼을 추가하세요.
4. 전입지별이 "서울특별시"인 행의 변화구분 값을 "자기지역"으로 수정하세요.
5. 전입지별, 2010, 2017, 증감_2010_2017, 변화구분 컬럼만 출력하세요.
---

```python
# 1. 전출지별이 "서울특별시"인 데이터만 추출한 후 복사본을 만드세요.
seoulDF = df[ df['전출지별'] == '서울특별시' ].copy()
```

```python

# 2. 2017년 이동자수에서 2010년 이동자수를 뺀 값을 "증감_2010_2017" 컬럼으로 추가하세요.
seoulDF["증감_2010_2017"] = seoulDF['2017'] - seoulDF['2010']
seoulDF.head(2)
```

```python
# 3. "증감_2010_2017" 값이 양수이면 "증가", 0이면 "동일", 음수이면 "감소"로 표시하는 "변화구분" 컬럼을 추가하세요.

# seoulDF["변화구분"] = np.where(
#     seoulDF["증감_2010_2017"] > 0, "증가",
#     np.where(seoulDF["증감_2010_2017"] == 0, "동일", "감소")
# )

seoulDF["변화구분"] = "동일"
seoulDF["변화구분"] = seoulDF["변화구분"].mask( seoulDF["증감_2010_2017"] > 0,  "증가" )
seoulDF["변화구분"] = seoulDF["변화구분"].mask(seoulDF["증감_2010_2017"] < 0,  "감소" )
```

```python
# 4. 전입지별이 "서울특별시"인 행의 변화구분 값을 "자기지역"으로 수정하세요.
seoulDF.loc[seoulDF["전입지별"] == "서울특별시", "변화구분"] = "자기지역"
```

```python
# 5. 전입지별, 2010, 2017, 증감_2010_2017, 변화구분 컬럼만 출력하세요.
result = seoulDF[["전입지별", "2010", "2017", "증감_2010_2017", "변화구분"]]
display(result)
```
