```python
## 데이터 준비 및 DF 인스턴스 생성
DATA_FILE = '../../DATA/iris.csv'

## 첫번째 줄 : 컬럼명 OK, 구분자 : 쉼표/콤마 OK
irisDF = pd.read_csv(DATA_FILE)
```

- 방법1) variety 컬럼을 object ==> 숫자/정수 라벨 인코딩


```python
## 품종명에 설정될 값 맵핑

irisDF_map = {
    "Setosa": 0,
    "Versicolor": 1,
    "Virginica": 2
}

## 결측값 NaN에는 함수 적용하여 NaN , 결과를 새로운 컬럼 추가
irisDF["variety_label"] = irisDF['variety'].map(irisDF_map)

## 확인
irisDF.head(3)
```

- 방법2) 특정 값만 1로 설정하는 인코딩 - 원-핫 인코딩

```python
irisDF2 = pd.get_dummies(irisDF,columns=["variety"],dtype=int)

irisDF2.tail()
```
